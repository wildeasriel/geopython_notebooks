### Functions

You can use Python for more advanced math by using a *{term}`function`*. Functions are pieces of code that perform a single action such as printing information to the screen (e.g., the `print()` function). Functions exist for a huge number of operations in Python.

Let's try out a few simple examples using functions to find the sine or square root of a value using the `sin()` and `sqrt()` functions.

```python
#This imports the various Python libraries used in this Notebook, with standard abbreviations used where necessary.
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#This uses the math.sin function to find the value of sine for a given angle. Python uses radians rather than degrees for
#angles, so to find the value of sine for an angle of 45º for example, we need to use an additional function, math.radians, to
#convert our degree value into radians. Typing math.sin(45) will print the value of 45 radians. Given that 1 radian = 57º,
#this will produce a highly inaccurate value.

math.sin(math.radians(45))
```

```python
#The code below is used to plot the value of sine from 0º to 360º using the Matplotlib.pyplot (Matplot) library, imported as
#plt. We first produce two empty lists, for our angles, and the value of sine for a given angle:

angles = []
sine_values = []

#We are dealing with 360 values here, so a for loop is appropriate to avoid calculating each value manually. The first line 
#specifies that we need to deal with each value from 0 to 360 by intervals of 1. The colon commands that the following indented
#lines of code should be performed for each of these values. We define each of these values with the new variable, 'angle'.

#We can suffix our two lists above (angles and sine_values) with .append. The first indented line populates the 'angles' list)
#with each value from 1 to 360. The second indented line uses the code in box 17 to calculate sine for each angle value. 

for angle in range(1,361):
    angles.append(angle)
    sine_values.append(math.sin(math.radians(angle)))

#We have therefore created two lists: A list of angles, and a sine value for each angle. These can be used to produce a graph.

#We use the plt library to do this. 

plt.plot(angles, sine_values, label="sin(θ)")
plt.xlabel("Angle(º)")
plt.ylabel("Sine value")
plt.title ("Sine values from 0º to 360º")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="red", linestyle="--", label="180º (zero crossing)")
plt.legend()
plt.grid()
plt.show()
```

### Math operations

: _**Table 2.1**. Basic math operations in Python._

|Operation      | Symbol | Example syntax | Returned value |

```python
math.sin(math.radians(3))
```

```python
math.sqrt(32/2)
```

Let's summarize what we've just seen with modules:

1. A {index}`module` is a group of code items such as functions that are related to one another. Individual modules are often in a group referred to as a library.

2. Modules can be loaded using the `import` statement. Functions that are part of the module `modulename` can then be used by typing `modulename.functionname()`. For example, `sin()` is a function that is part of the `math` module, and used by typing `math.sin()` with some number between the parentheses.

3. In a Jupyter Notebook the variables you define earlier code cells will be available for use in the cells that follow as long as you have already executed the cells.

Note that modules may also contain constants such as `math.pi`. Parentheses are not used when calling constant values.

## Data types
